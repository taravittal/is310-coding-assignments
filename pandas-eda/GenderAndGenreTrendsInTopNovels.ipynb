# Gender and Genre Trends in Top Novels

This notebook explores how genre, publication year, ranking, and pronoun counts reveal patterns in the Goodreads Top Ranked Novels and New York Times Bestseller datasets.

## Loading the Data

```python
import pandas as pd
import altair as alt

combined_novels_nyt_df = pd.read_csv('combined_novels_viz.csv')
```

## Data Overview

```python
combined_novels_nyt_df.head()
combined_novels_nyt_df.info()
print(combined_novels_nyt_df.head())
print(combined_novels_nyt_df.shape)
print(combined_novels_nyt_df.columns)
```

## Genre Patterns

```python
alt.Chart(combined_novels_nyt_df).mark_bar().encode(
    x=alt.X('genre:N', title='Genre'),
    y=alt.Y('count():Q', title='Number of Novels'),
    tooltip=['genre:N', 'count():Q']
).properties(
    title='Number of Novels by Genre',
    width=600,
    height=400
)
print(combined_novels_nyt_df['genre'].value_counts())
```

This graph shows how many novels appear in each genre. Some genres appear much more often than others, which suggests the dataset is not evenly balanced. This matters because overrepresented genres may shape the overall patterns we find.

## Ranking Patterns by Genre

```python
alt.Chart(combined_novels_nyt_df).mark_bar().encode(
    x=alt.X('genre:N', title='Genre'),
    y=alt.Y('mean(top_500_rank):Q', title='Average Top 500 Rank'),
    tooltip=['genre:N', 'mean(top_500_rank):Q']
).properties(
    title='Average Top 500 Rank by Genre',
    width=600,
    height=400
)
print(combined_novels_nyt_df.groupby('genre')['top_500_rank'].mean())
```

This graph compares the average ranking for each genre. A lower average rank means books in that genre tend to appear closer to the top of the list. This helps show whether certain genres are ranked more favorably than others.

## Pronoun Counts by Genre

```python
melted_df = pd.melt(
    combined_novels_nyt_df,
    id_vars=['title', 'author', 'pub_year', 'genre'],
    value_vars=['he_counts', 'she_counts', 'they_counts'],
    var_name='pronoun',
    value_name='pronoun_count'
)

melted_df['pronoun'] = melted_df['pronoun'].str.replace('_counts', '')

alt.Chart(melted_df[melted_df['genre'].notna()]).mark_bar().encode(
    x=alt.X('genre:N', title='Genre'),
    y=alt.Y('sum(pronoun_count):Q', title='Total Pronoun Count'),
    color='pronoun:N',
    tooltip=['genre:N', 'pronoun:N', 'sum(pronoun_count):Q']
).properties(
    title='Pronoun Counts by Genre',
    width=700,
    height=400
)
print(melted_df.head())
print(melted_df.groupby('pronoun')['pronoun_count'].sum())
```

This graph compares the total counts of he, she, and they across genres. Pronoun counts can help reveal patterns in representation, though they do not fully explain gender or identity in the novels. These results should be treated as a starting point for more careful interpretation.

## Genre Trends Over Time

```python
combined_novels_nyt_df['pub_date'] = pd.to_datetime(
    combined_novels_nyt_df['pub_year'].astype(str) + '-01-01',
    errors='coerce'
)

alt.Chart(combined_novels_nyt_df[combined_novels_nyt_df['genre'].notna()]).mark_bar().encode(
    x=alt.X('pub_date:T', title='Publication Year'),
    y=alt.Y('count():Q', title='Number of Novels'),
    color='genre:N',
    tooltip=['genre:N', 'count():Q']
).properties(
    title='Top 500 Novels by Genre Over Time',
    width=800,
    height=400
)
print(combined_novels_nyt_df['pub_year'].describe())
```

This graph shows how genres appear across publication years. It can help identify whether certain genres are more common in particular historical periods. Since the dataset only includes top-ranked novels, these trends also reflect the choices behind the ranking system.

## Limitations and Biases

This dataset is useful for exploring literary patterns, but it is not neutral. Some genres, authors, languages, and time periods may be overrepresented. Missing values and ranking systems can also shape what patterns appear in the analysis.

## Conclusion

This EDA suggests that genre, publication year, and pronoun use can reveal important patterns in the Top 500 Novels dataset. The data also has limitations because some groups and categories appear more often than others. Overall, this notebook shows how exploratory data analysis can help create future research questions rather than prove one final answer.
